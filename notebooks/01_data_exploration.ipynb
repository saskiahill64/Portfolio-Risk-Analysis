```python
#--------------------------------------------------
# Imports
#--------------------------------------------------
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
#-----------------------------------------------------------------------------
# Calculating the number of trading days in each month of 2020 for AAPL stock
#-----------------------------------------------------------------------------

jan_data = yf.download("AAPL", start="2020-01-01", end="2020-02-01") # downloading the data from the yfinance module for January 2020
jan_trading_days = jan_data.shape[0] # storing the number of rows for January 2020

feb_data = yf.download("AAPL", start="2020-02-01", end="2020-03-01")
feb_trading_days = feb_data.shape[0]

mar_data = yf.download("AAPL", start="2020-03-01", end="2020-04-01")
mar_trading_days = mar_data.shape[0]

apr_data = yf.download("AAPL", start="2020-04-01", end="2020-05-01")
apr_trading_days = apr_data.shape[0]

may_data = yf.download("AAPL", start="2020-05-01", end="2020-06-01")
may_trading_days = may_data.shape[0]

jun_data = yf.download("AAPL", start="2020-06-01", end="2020-07-01")
jun_trading_days = jun_data.shape[0]

jul_data = yf.download("AAPL", start="2020-07-01", end="2020-08-01")
jul_trading_days = jul_data.shape[0]

aug_data = yf.download("AAPL", start="2020-08-01", end="2020-09-01")
aug_trading_days = aug_data.shape[0]    

sep_data = yf.download("AAPL", start="2020-09-01", end="2020-10-01")
sep_trading_days = sep_data.shape[0]

oct_data = yf.download("AAPL", start="2020-10-01", end="2020-11-01")
oct_trading_days = oct_data.shape[0]

nov_data = yf.download("AAPL", start="2020-11-01", end="2020-12-01")
nov_trading_days = nov_data.shape[0]

dec_data = yf.download("AAPL", start="2020-12-01", end="2021-01-01")
dec_trading_days = dec_data.shape[0]

trading_days_2020 = pd.DataFrame({"Month": ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"],
                                   "Trading Days": [jan_trading_days, feb_trading_days, mar_trading_days, apr_trading_days, may_trading_days, jun_trading_days, jul_trading_days, aug_trading_days, sep_trading_days, oct_trading_days, nov_trading_days, dec_trading_days]})
print(trading_days_2020)

total_trading_days = trading_days_2020["Trading Days"].sum()
print(f"Total trading days in 2020: {total_trading_days}")
```

### Number of Trading Days

The above code calculates a total of 253 trading days in 2020, which is less than the 365 calendar days because the stock markets are only open on weekdays and also close on public hoidays.

```python
#-----------------------------------------------------------------------------
# Downloading Apple and Microsoft Data 
# Checking for missing values
#-----------------------------------------------------------------------------

tickers = [
    "SPY",   # US equities
    "QQQ",   # Nasdaq
    "GLD",   # Gold
    "TLT",   # US Treasury bonds
    "VNQ",   # Real estate
    "USO"    # Oil
]
prices = yf.download( # Download historical Apple and Microsoft stock prices
    tickers,
    start="2020-01-01",
    end="2025-01-01")

print(prices.isna().sum()) # Sum up any missing values
```

### Checking for Missing Values

Historical data from Apple and Microsoft has been downloaded, and the sum of any missing values was calculated to be zero. Therefore, the data from the chosen time period are complete and no further data cleaning is required before analysis.

```python
#-----------------------------
# Basic statistics
#-----------------------------

#print(prices.info())

print(prices.describe())
```

### Comparing Microsoft and Apple Stock Prices

The above data exploration demonstrates that Nasdaq (QQQ) has highest mean closing stock price over the time period anaylsed compared to the rest of the stocks.

It is important to note that although QQQ had higher mean stock prices, this does not necessarily mean that Microsoft had a higher market capitalisation, as this depends on the number of shares available:

Market cap = share price x number of outstanding shares


```python
#-----------------------------------------------------------------------------
# Calculate daily returns
#-----------------------------------------------------------------------------

daily_returns = prices['Close'].pct_change() * 100  # Calculate daily returns as percentage change

#print(daily_returns.head())

print(daily_returns.describe())
```

### Calculating and Comparing Daily Returns

Above the average daily percentage change in stock prices over the time period has been calculated. This is called the return of a stock, and allows the performance of different stocks to be compared regardless of their starting prices. The values for both companies are similar, but QQQ has the highest mean daily return. Oil (USO) has the highest standard deviation of daily returns, indicating that the oil stocks were the most volatile.


```python
#-----------------------------------------------------------------------------
# Plotting the stock prices of AAPL and MSFT
#-----------------------------------------------------------------------------

plt.plot(prices['Close']['SPY'], label='SPY')
plt.plot(prices['Close']['GLD'], label='GLD')
plt.plot(prices['Close']['QQQ'], label='QQQ')
plt.plot(prices['Close']['TLT'], label='TLT')
plt.plot(prices['Close']['VNQ'], label='VNQ')
plt.plot(prices['Close']['USO'], label='USO')
plt.title('Stock Prices of SPY, GLD, QQQ, TLT, VNQ, and USO')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

```

```python
#-----------------------------------------------------------------------------
# Plotting the daily returns of AAPL and MSFT
#-----------------------------------------------------------------------------

plt.plot(daily_returns['SPY'], label='SPY')
plt.plot(daily_returns['GLD'], label='GLD')
plt.plot(daily_returns['QQQ'], label='QQQ')
plt.plot(daily_returns['TLT'], label='TLT')
plt.plot(daily_returns['VNQ'], label='VNQ')
plt.plot(daily_returns['USO'], label='USO')
plt.title('Daily Returns of SPY, GLD, QQQ, TLT, VNQ, and USO')
plt.xlabel('Date')
plt.ylabel('Daily Returns (%)')
plt.legend()
plt.show()
```

```python
#-----------------------------------------------------------------------------
# Calculating volatility (standard deviation of daily returns) for AAPL and MSFT
#-----------------------------------------------------------------------------

volatility = daily_returns.std()  # Calculate the standard deviation of daily returns

print(volatility)
```

```python
#-----------------------------------------------
# Correlation Matrix
#------------------------------------------------

correlation = daily_returns.corr()
print(correlation)

corr_matrix = daily_returns.corr(method="pearson")

fig, ax = plt.subplots(figsize=(8, 6))

cax = ax.imshow(
    corr_matrix,
    cmap="coolwarm",
    vmin=-1,
    vmax=1
    )

cbar = fig.colorbar(cax, ax=ax)
cbar.set_label("Pearson correlation coefficient", rotation=270, labelpad=15)

labels = corr_matrix.columns

ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))

ax.set_xticklabels(labels, rotation=45, ha="right")
ax.set_yticklabels(labels)

for i in range(len(labels)):
    for j in range(len(labels)):
        ax.text(
        j, i,
        f"{corr_matrix.iloc[i, j]:.2f}",
        ha="center",
        va="center",
        fontsize=9,
        color="black"
    )
ax.set_title("Correlation Matrix", pad=15)
ax.grid(False)

plt.show()
```

### Plotting a Correlation Matrix

The correlation matrix above shows the pearson correlation coefficients between the average daily returns of each stock. It can be seen that there is a very strong positive correlation between QQQ and SPY. This makes sense since SPY is a stock index tracking the performance of the US's top 500 companies and QQQ (Nasdaq) is also a stock index which lists many of the same companies as SPY. Therefore, the strong correlation of 0.93 between SPY and QQQ is expected.
